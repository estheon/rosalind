```python
# Load important packages
import Bio
from Bio import SeqIO, SearchIO, Entrez
from Bio.Seq import Seq
from Bio.SeqUtils import GC
from Bio.Blast import NCBIWWW
from Bio.Data import CodonTable
```

```python
RNA = open("rosalind_prot.txt","r")
RNA = Seq(RNA.read().rstrip())
print(RNA)
print(type(RNA))
```

```python
prot = RNA.translate()
print(prot)
print(type(prot))
# Asterisk in the sequence represents stop codon, which must be dropped
```

```python
# Turn it back into a string for string manipulation
prot_output = str(prot)
prot_output = prot_output[:-1]
print(prot_output)
```

```python
o = open("PROT_o.txt","w")
o.write(prot_output)
o.close()
```
